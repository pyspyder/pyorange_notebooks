```python
data = data_set.sample(frac=0.97, random_state=786)
data_unseen = data_set.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

# Initializing the Experiment

```python
#import classification module
from pyorange.classification import *
#intialize the setup
clf = setup(data, target = 'Purchase')
```

# comparing all the models and auto selecting the best model based on accuracy

```python
best_model = compare_models()
```

# saving the trained metrics of different models for reporting

```python
metrics = pull()
metrics.to_csv('metrics.csv')
print(metrics)
```

# Printing the best model

```python
print(best_model)
```

# printing all the available models

```python
models()
```

# creating decision tree model

```python
dt = create_model('dt')
```

# creating random forest model

```python
rf = create_model('rf')
```

# tuning the decision tree model

```python
tuned_dt = tune_model(dt)
```

# tuning the random forest model

```python
tuned_rf = tune_model(rf)
```

# Tuning the best model

```python
tuned_best = tune_model(best_model)
```

```python
plot_model(tuned_best, plot = 'confusion_matrix')
```

# evaluating the model

```python
evaluate_model(tuned_best)
```

```python
predict_model(tuned_best);
```

# finalizing the tuned best model

```python
final_best = finalize_model(tuned_best)
```

# saving the best model

```python
save_model(final_best,'one_vs_all_best')
```
